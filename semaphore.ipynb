# Semaphore API integration tests

This notebook runs Semaphore's public/user-visible APIs, such as listing broadcasts, to verify that the service is running.

```python
from lsst.rsp import RSPClient
```

```python
client = RSPClient("semaphore")
```

## Access Semaphore app metadata

```python
r = await client.get("/")
metadata = r.json()
print(f"Semaphore version {metadata['metadata']['version']}")
```

## Check broadcasts endpoint

```python
r = await client.get("/v1/broadcasts")
broadcasts = r.json()
assert isinstance(broadcasts, list)
```
