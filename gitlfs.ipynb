```python
import subprocess
from tempfile import TemporaryDirectory
from pathlib import Path

with TemporaryDirectory() as td:
    
    repo="lsst/tutorial-notebooks-data"
    gh_url=f"https://github.com/{repo}"
    t_repo = Path(td) / "tutorial-notebooks-data"

    testfile = "PGC_038749_I_g_bbl2011.fits"

    cmd=[ "git", "clone", gh_url, str(t_repo) ]
   
    subprocess.run(cmd)

    tf = t_repo / "data" / testfile

    with tf.open("rb") as f:
        sixbytes = f.read(6)
        sixstr = sixbytes.decode()

        # If it IS a FITS file, we get this...
        # If not it will be the Git LFS header
        assert sixstr == "SIMPLE"

```
