# Git LFS test

This notebook checks out a git-lfs-backed notebook and verifies that it retrieved the (LFS-stored) contents, rather than the Git LFS header.

## Imports and constants

```python
import subprocess
from tempfile import TemporaryDirectory
from pathlib import Path

repo="tutorial-notebooks-data"
gh_url=f"https://github.com/lsst/{repo}"
testfile = "PGC_038749_I_g_bbl2011.fits"
```

## Clone repo and read first six bytes of test file

```python
with TemporaryDirectory() as td:
    t_repo = Path(td) / repo

    cmd=[ "git", "clone", gh_url, str(t_repo) ]  
    subprocess.run(cmd)

    tf = t_repo / "data" / testfile
    with tf.open("rb") as f:
        sixbytes = f.read(6)
```

## Ensure that those bytes are the file contents, and not the LFS header

```python
sixstr = sixbytes.decode()
assert sixstr == "SIMPLE", f"Expected 'SIMPLE', received '{sixstr}'"
```
