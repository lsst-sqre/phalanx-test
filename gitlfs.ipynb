# Git LFS test

This notebook checks out a git-lfs-backed notebook and verifies that it retrieved the (LFS-stored) contents, rather than the Git LFS header.

## Imports and constants

```python
import shutil
import subprocess
from tempfile import TemporaryDirectory
from pathlib import Path

repo="tutorial-notebooks-data"
gh_url=f"https://github.com/lsst/{repo}"
testfile = "PGC_038749_I_g_bbl2011.fits"
```

## Clone repo and verify first six bytes of test file

```python
repo_path=Path(repo)
if repo_path.is_dir():
    shutil.rmtree(repo_path)

cmd=[ "git", "clone", gh_url, str(repo_path) ]  
proc=subprocess.run(cmd, capture_output=True)

tf = repo_path / "data" / testfile
f = tf.open("rb")
sixbytes = f.read(6)
f.close()
sixstr = sixbytes.decode()
assert sixstr == "SIMPLE", f"Expected 'SIMPLE', received '{sixstr}'; stdout={proc.stdout.decode()},stderr={proc.stderr.decode()}"
```

## Clean up directory

```python
shutil.rmtree(repo_path)
```
